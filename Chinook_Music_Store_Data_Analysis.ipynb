**Part 2: Transformation**

**Retrieves specific information about tracks with a particular name and their associated media types.**

The addition of the 'Media Type' column to Tracks table. This transformation involves combining information from two separate tables (tracks and media_types) into a single result set by joining them based on a common condition (track_name)

```python
# Join Track and MediaType
sq('''select
t.name as track_name ,
t.Composer,
Milliseconds as duration_in_ms,
t.UnitPrice ,
m.Name as media_type
from tracks as t
join media_types as m
where track_name="For Those About To Rock (We Salute You)"
''').rename(columns = lambda x: x.replace('_',' ').upper())
```
